```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
loader = PyPDFLoader("budget_speech.pdf")
```

```python
loader
```

```python
pages = loader.load()
```

```python
pages
```

```python
pages[0]
```

```python
len(pages) #length of pages
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    length_function=len
)   
```

```python
chunks = text_splitter.split_documents(pages)
```

```python
# vector databases to store numbers
! pip install faiss-cpu
```

```python
# installing the sentence transformer model from hugging face
! pip install sentence-transformers
```

```python
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
embedding_model_name = "all-MiniLM-L6-v2"
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
vector_db = FAISS.from_documents(chunks, embeddings)
```

```python
result = vector_db.similarity_search(query="Building Rural Prosperity and Resilience?")
```

```python
result1 = vector_db.similarity_search(query="What is the budget allocation for healthcare?")
```
